### **Data Cleaning**

```python
df.isna().sum()
```

#### **Check for missing Values**

```python
# Check percentage of missing values per column
missing_values = df.isnull().mean().sort_values(ascending=False) * 100
missing_values[missing_values > 0]
```

#### **Drop Columns with too many missing values**

```python
# Drop columns with more than 50% missing values
df_cleaned = df.loc[:, df.isnull().mean() < 0.5]
print("Remaining columns after dropping high-missing ones:", df_cleaned.shape[1])
```

#### **Key Columns of interest**

```python
# Choose only important columns
columns_of_interest = [
    'date', 'continent', 'location', 'total_cases', 'total_deaths',
    'new_cases', 'new_deaths', 'total_vaccinations',
    'people_fully_vaccinated_per_hundred'
]

df = df[columns_of_interest]
```

```python
df.head()
```

#### **Filter Countries of Interest**

```python
countries = ['Kenya', 'United States', 'India','China', 'United Kingdom', 'Italy', 'South Africa', 'Germany', 'Australia']
df_selected = df[df['location'].isin(countries)]
```

#### **Drop Rows with missing date, location**

```python
df = df.dropna(subset=['date', 'location'])
```

#### **Convert date Column to Datetime Format**

```python
df['date'] = pd.to_datetime(df['date'])
```

#### **Sort Data by Country and Date**

```python
df = df.sort_values(by=['location', 'date'])
```

#### **Fill Missing Numeric Values**

```python
# Fill missing values using interpolation for each country
df[['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations', 'people_fully_vaccinated_per_hundred']] = \
    df.groupby('location')[['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations', 'people_fully_vaccinated_per_hundred']].transform(lambda group: group.interpolate())

# Optionally, fill any remaining missing values with 0
df.fillna(0, inplace=True)
```

### **Exploratory Data Analysis (EDA)**

#### **Plot total cases over time for selected countries**

```python
# Plot total cases over time
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df_selected[df_selected['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### **Plot total deaths over time**

```python
#Plot total deaths over time
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df_selected[df_selected['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.title('Total Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### **Compare daily new cases between countries**

```python
#Daily new cases comparison
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df_selected[df_selected['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.title('Daily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### **The death rate: total_deaths / total_cases**

```python
df_selected = df_selected.copy()  # Avoid SettingWithCopyWarning
df_selected['death_rate'] = df_selected['total_deaths'] / df_selected['total_cases']

plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df_selected[df_selected['location'] == country]
    plt.plot(country_data['date'], country_data['death_rate'], label=country)

plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### **Top 10 Countries by Total Cases (Latest Date)**

```python
# Load data and convert date column
df = pd.read_csv("/Users/Jumah/Downloads/cleaned_covid_data.csv")
df['date'] = pd.to_datetime(df['date'])

# Remove rows with missing total_cases or country names
df_clean = df[['location', 'date', 'total_cases']].dropna()
df_clean = df_clean[~df_clean['location'].str.startswith('World') & ~df_clean['location'].isin(['International'])]

# Keep only the latest date for each country
latest_cases = df_clean.sort_values('date').groupby('location').tail(1)

# Get top 10 countries by total cases
top_10 = latest_cases.sort_values(by='total_cases', ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(top_10['location'], top_10['total_cases'], color='#B22222')
plt.title('Top 10 Countries by Total COVID-19 Cases (Latest Available)')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')

# Add numbers on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height):,}', 
             ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.show()
```

### **Chloropleth Map**

```python
# Filter relevant columns and drop missing values
choropleth_df = df[['location', 'iso_code', 'date', 'total_cases']].dropna(subset=['iso_code', 'total_cases'])

# Keep the latest record per country
choropleth_df = choropleth_df.sort_values('date').groupby('iso_code', as_index=False).last()

# Optional: Remove non-country entities (e.g., continents, income levels)
choropleth_df = choropleth_df[choropleth_df['iso_code'].str.len() == 3]
```

##### **Plot the Chloropleth Map**

```python
# Create the choropleth map
fig = px.choropleth(
    choropleth_df,
    locations="iso_code",  # ISO Alpha-3 codes
    color="total_cases",   # What to color by
    hover_name="location", # Tooltip country name
    color_continuous_scale="OrRd",
    title="Total COVID-19 Cases by Country (Latest Available Date)"
)

fig.update_layout(geo=dict(showframe=False, showcoastlines=False))
fig.show()
```
